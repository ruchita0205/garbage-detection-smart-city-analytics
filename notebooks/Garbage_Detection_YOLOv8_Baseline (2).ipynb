dataset verification

```python
import os

base_path = "/content/drive/MyDrive/Garbage_Detection_Smart_City_Project/Dataset/Roboflow_Dataset/Garbage Detection.v21i.yolov8"

for split in ["train", "valid", "test"]:
    img_path = os.path.join(base_path, split, "images")
    lbl_path = os.path.join(base_path, split, "labels")

    images = len(os.listdir(img_path))
    labels = len(os.listdir(lbl_path))

    print(f"{split.upper()}")
    print("Images:", images)
    print("Labels:", labels)
    print("-"*30)
```

class names

```python
import yaml

yaml_path = f"{base_path}/data.yaml"

with open(yaml_path, "r") as f:
    data = yaml.safe_load(f)

print("Number of Classes:", data["nc"])
print("Class Names:")
print(data["names"])
```

checking missing labels

```python
import os

missing_labels = []

img_dir = f"{base_path}/train/images"
label_dir = f"{base_path}/train/labels"

for img in os.listdir(img_dir):
    label = os.path.splitext(img)[0] + ".txt"

    if not os.path.exists(os.path.join(label_dir, label)):
        missing_labels.append(img)

print("Missing Labels:", len(missing_labels))
```

```python
import os

img_dir = f"{base_path}/train/images"
label_dir = f"{base_path}/train/labels"

missing = []

for img in os.listdir(img_dir):

    label = os.path.splitext(img)[0] + ".txt"

    if not os.path.exists(os.path.join(label_dir, label)):
        missing.append(img)

print("Missing:", len(missing))

print("\nFirst 10 Missing Files:\n")

for x in missing[:10]:
    print(x)
```

```python
import os

missing = []

img_dir = f"{base_path}/train/images"
label_dir = f"{base_path}/train/labels"

for img in os.listdir(img_dir):
    label = os.path.splitext(img)[0] + ".txt"

    if not os.path.exists(os.path.join(label_dir, label)):
        missing.append(img)

duplicates = [x for x in missing if "(1)" in x]

print("Total Missing:", len(missing))
print("Duplicate-style Missing:", len(duplicates))
```

class distribution

```python
from collections import Counter
import os

counter = Counter()

label_dir = f"{base_path}/train/labels"

for file in os.listdir(label_dir):

    if file.endswith(".txt"):

        with open(os.path.join(label_dir, file), "r") as f:

            for line in f.readlines():

                cls = int(line.split()[0])

                counter[cls] += 1

print(counter)
```

```python
for k,v in counter.items():
    print(data["names"][k], v)
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

results = model.train(
    data="/content/drive/MyDrive/Garbage_Detection_Smart_City_Project/Dataset/Roboflow_Dataset/Garbage Detection.v21i.yolov8/data.yaml",
    epochs=30,
    imgsz=640,
    batch=8,
    patience=10,
    optimizer="AdamW",
    lr0=0.001,
    cos_lr=True,
    name="garbage_yolov8n_improved_30epochs"
)
```
