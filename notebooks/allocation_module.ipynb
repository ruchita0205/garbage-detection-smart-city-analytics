```python
import pandas as pd

hotspots = pd.DataFrame({
    "Cluster": [0, 1, 2],
    "SWSI": [79.11, 45.31, 20.70],
    "Priority": [1, 2, 3]
})

def allocate_resources(swsi):
    if swsi >= 70:
        workers = 6
        vehicles = 3

    elif swsi >= 40:
        workers = 4
        vehicles = 2

    elif swsi >= 20:
        workers = 2
        vehicles = 1

    else:
        workers = 1
        vehicles = 1

    return workers, vehicles

hotspots[["Workers", "Vehicles"]] = hotspots["SWSI"].apply(
    lambda x: pd.Series(allocate_resources(x))
)

print(hotspots)

hotspots.to_excel(
    "allocation_output.xlsx",
    index=False
)

print("\nExcel file saved successfully.")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(
    hotspots["Cluster"].astype(str),
    hotspots["Workers"]
)

plt.title("Worker Allocation by Hotspot Cluster")
plt.xlabel("Cluster")
plt.ylabel("Number of Workers")

plt.grid(axis='y')

plt.show()
```

```python
plt.savefig("worker_allocation.png", bbox_inches="tight")
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    hotspots["Cluster"].astype(str),
    hotspots["Vehicles"]
)

plt.title("Vehicle Allocation by Hotspot Cluster")
plt.xlabel("Cluster")
plt.ylabel("Number of Vehicles")

plt.grid(axis='y')

plt.show()

plt.savefig("vehicle_allocation.png", bbox_inches="tight")
```
